# Comparing yields depending on round and acquisition type

```python
rxns_yields_rounds = rxns_df[["Yield of 3aa","Round"]]
rxns_conv_rounds = rxns_df[["Conversion of 1a", "Round"]]
```

```python
rxns_yields_rounds_TS = rxns_yields_rounds.iloc[:88,:]
rxns_conv_rounds_TS = rxns_conv_rounds.iloc[:88,:]
```

```python
sns.scatterplot(data=rxns_yields_rounds_TS, x="Round", y="Yield of 3aa")
print(np.max(rxns_yields_rounds_TS["Yield of 3aa"]))
```

```python
sns.scatterplot(data=rxns_conv_rounds_TS, x="Round", y="Conversion of 1a")
```

## EI -> EDBO with Expected Improvement

```python
rxns_yields_rounds_EI = rxns_yields_rounds.iloc[89:176,:]
rxns_conv_rounds_EI = rxns_conv_rounds.iloc[89:176,:]
```

```python
sns.scatterplot(data=rxns_yields_rounds_EI, x="Round", y="Yield of 3aa")
print(np.max(rxns_yields_rounds_EI["Yield of 3aa"]))
```

```python
sns.scatterplot(data=rxns_conv_rounds_EI, x="Round", y="Conversion of 1a")
```

## UCB -> Automus with Upper Confidence Bound

```python
rxns_yields_rounds_UCB = rxns_yields_rounds.iloc[176:,:]
rxns_conv_rounds_UCB = rxns_conv_rounds.iloc[176:,:]
```

```python
sns.scatterplot(data=rxns_yields_rounds_UCB, x="Round", y="Yield of 3aa")
print(np.max(rxns_yields_rounds_UCB["Yield of 3aa"]))
```

```python
sns.scatterplot(data=rxns_conv_rounds_EI, x="Round", y="Conversion of 1a")
```

The EI acquisition type leads to the highest yield

Note that the higher the conversion, the higher the yield (very much obvious). The opposite isn't necessarily true.

# Comparing yields depending on the temperature

```python
rxns_yield_60 = []
for i in range (0,len(rxns_df["T [°C]"])):
    if rxns_df["T [°C]"][i]==60:
        rxns_yield_60.append(rxns_df["Yield of 3aa"][i])
rxns_yield_80 = []
for i in range (0,len(rxns_df["T [°C]"])):
    if rxns_df["T [°C]"][i]==80:
        rxns_yield_80.append(rxns_df["Yield of 3aa"][i])

plt.pyplot.plot(rxns_yield_60, label='T=60')
plt.pyplot.plot(rxns_yield_80, label= 'T=80')
plt.pyplot.legend()
```

More reactions took place at 80°C, and they overall produced a higher yield than at 60°C.
